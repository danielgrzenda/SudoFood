```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
ID = '61b7fa8a'
KEY = '10dc6fd573524c488bcef197bcebaaf4'

MATT_ID = '8738e7a2'
MATT_KEY = '234f4fe2774becb61689975054fb6d03'

def recipe_search(max_result=None,start=None):
    if(max_result == None and start==None):
        response = requests.get('http://api.yummly.com/v1/api/recipes?_app_id=%s&_app_key=%s'%(ID,KEY)).json()
    elif(max_result != None and start==None):
        response = requests.get('http://api.yummly.com/v1/api/recipes?_app_id=%s&_app_key=%s&maxResult=%s'%(ID,KEY,max_result)).json()
    else:
        response = requests.get('http://api.yummly.com/v1/api/recipes?_app_id=%s&_app_key=%s&maxResult=%s&start=%s'%(ID,KEY,max_result,start)).json()
    return response

def get_one_recipe(recipe_id):
    response = requests.get('http://api.yummly.com/v1/api/recipe/%s?_app_id=%s&_app_key=%s'%(recipe_id,ID,KEY))
    return response
```

```python
recipe_list = set()
```

```python
for i in range(0,500000,1000):
    try:
        data = recipe_search(max_result=1000,start=i)
    except (KeyboardInterrupt, SystemExit):
        raise
    except:
        continue
    for item in data['matches']:
        recipe_list.add(item['id'])
    print(len(recipe_list))
```
